```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for anomalies
regions = [
    "Bni Mellal-Khnifra",
    "Casablanca-Settat",
    "Dakhla-Oued Ed-Dahab",
    "Dra-Tafilalet",
    "Fs-Mekns",
    "Guelmim-Oued Noun",
    "Layoune-Sakia El Hamra",
    "Marrakech-Safi",
    "Oriental",
    "Rabat-Sal-Knitra",
    "Souss-Massa",
    "Tanger-Tetouan-Al Hoceima",
]

anomalies = [
    12, 2, 6, 12, 6, 10, 10, 10, 4, 3, 7, 8
]

data = pd.DataFrame({"Region": regions, "Anomalies": anomalies})

data = data.sort_values(by="Anomalies", ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(data["Region"], data["Anomalies"], color="skyblue")
plt.xlabel("Nombre d'Anomalies", fontsize=12)
plt.ylabel("Region", fontsize=12)
plt.title("Anomalies pour chaque region", fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis for better visual order
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
